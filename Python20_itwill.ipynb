# `matplotlib.pyplot` 모듈을 사용한 데이터 시각화

```python
# 순서
## 1. Figure 객체 생성
fig = plt.figure()

## 2. Figure 객체에 Subplot(들)을 추가
## 첫 번째 방법) axis = fig.add_subplot(1, 1, 1)     ### -> Figure.add_subplot(nrows, ncols, index)
### ->> nrows: subplot의 row(행)의 개수
### ->> ncols: subplot의 column(열)의 개수
### ->> index: subplot의 인덱스(위치)

axis = fig.add_subplot(111) ## 두 번째 방법: Figure.add_subplot(pos)
### ->> pos: '행/열/인덱스' 순서로 만들어진 3자리 출력

## 3. subplot에 그래프를 그림
axis.hist(x=tips.total_bill)

## 4. 그래프의 옵션들을 설정
axis.set_title('Histogram(total_bill)')
axis.set_xlabel('total_bill')
axis.set_ylabel('count')

## 5. 그래프 보여주기
plt.show()
```

```python
# 1. Figure 객체 생성 + Subplot 추가
fig, axis = plt.subplots()   ## subplots의 파라미터 nrow와 nclos의 기본값은 1

# 2. Subplot에 그래프 그림
axis.hist(x=tips.tip)

# 3. 그래프 옵션 설정
axis.set_xlabel('Tip')
axis.set_ylabel('Count')

# 4. 그래프 보여주기
plt.show()
```

1개의 Figure에 1개의 Subplot만 있는 경우에는 pyplot 모듈의 top-level 그래프 함수를 직접 이용할 수도 있음

top-level 그래프 함수를 사용하면 Figure 생성과 Subplot 추가 코드를 작성할 필요가 없음

```python
# pyplot 함수 호출 -> Figure 객체 생성 + Subplot(Axex) 객체 생성 + 그래프 생성
plt.hist(x = tips.total_bill)

# 그래프 옵션 설정 -> pyplot 모듈의 함수 사용
plt.xlabel('Total_bill')
plt.ylabel('Count')
plt.title('Histogram')

# 그래프 보여주기
plt.show()
```

* 1개의 Figure에 2개의 Subplot을 위/아래로 배치 (nrows 설정)
* 위쪽 Axex(축)에는 total_bill 히스토그램을 작성
* 아래쪽 Axex(축)에는 산점도 그래프(x축: total_bill / y축: tip)

```python
# 1. Figure 객체 생성
fig = plt.figure(figsize=(6, 10))

# 2. Figure 객체에 Subplot들을 추가
axis1 = fig.add_subplot(2, 1, 1)  ## -> 위쪽 axes, add_subplot(211)과 동일
axis2 = fig.add_subplot(212)  ## -> 아래쪽 axes

# 3. Subplot에 그래프 그림
axis1.hist(x=tips.total_bill)
axis2.scatter(x=tips.total_bill, y=tips.tip)

# 4. 그래프 옵션 설정
axis1.set_title('Histogram')
axis1.set_xlabel('Total Bill')
axis1.set_ylabel('Count')
axis2.set_title('Scatter plot')
axis2.set_xlabel('Tip')
axis2.set_ylabel('Count')

# 5. 그래프 보여주기
plt.show()
```

* 1개의 Figure에 2개의 Subplot을 좌/우로 배치 (ncols 설정)
* 왼쪽 Axex(축)에는 total_bill 히스토그램을 작성
* 오른쪽 Axex(축)에는 산점도 그래프(x축: total_bill / y축: tip)

```python
# 1. Figure 생성 + Subplot 추가
fig, axes = plt.subplots(ncols=2, figsize = (12, 4))   ## -> `plt.subplots() 사용하면 배열 형태의 여러개 axes 생성
# fig.set_figwidth(12)

# 2. 왼쪽 축(Axis)에 그래프 그리기 & 옵션 설정
axes[0].hist(x=tips.total_bill)
axes[0].set_title('Histogram')
axes[0].set_xlabel('Total_bill')
axes[0].set_ylabel('Count')

# 3. 오른쪽 축(Axis)에 그래프 그리기 & 옵션 설정
axes[1].scatter(x=tips.total_bill, y=tips.tip)
axes[1].set_title('Scatter')
axes[1].set_xlabel('Total_bill')
axes[1].set_ylabel('Tip')


# 4. 그래프 보여주기
plt.show()
```

```python
# 히스토그램 그리기: total_bill과 tip을 한 번에 리스트로 전달
plt.hist([tips.total_bill, tips.tip], label=["total_bill", "tip"])

plt.xlabel('')
plt.legend()                          # 범례(legend) 추가
plt.title("Two Histograms")           # 그래프 제목 설정
plt.show()                            # 그래프 화면에 출력
```

```python
plt.hist(tips.total_bill, alpha=0.5, label='total_bill')  # total_bill 히스토그램, 투명도 50%
plt.hist(tips.tip, alpha=0.5, label='tip')                # tip 히스토그램, 투명도 50%
plt.legend()                                              # 범례 표시
plt.title("Two Overlapping Histograms")                   # 제목
plt.show()                                                # 출력
```

```python
# 첫 번째 그래프 (왼쪽)
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째
plt.hist(tips.total_bill, color='skyblue')
plt.title("Total Bill")
plt.xlabel('total bill')
plt.ylabel('count')

# 두 번째 그래프 (오른쪽)
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째
plt.hist(tips.tip, color='salmon')
plt.title("Tip")
plt.xlabel('tip')
plt.ylabel('count')

plt.tight_layout()  # 레이아웃 자동 정리 (그래프 겹치지 않게)
plt.show()
```

## numpy.ndarray 클래스 타입 vs Python list 클래스 타입
* list는 `list + list` 연산과 `list * int` 연산만 제공
* numpy.ndarray는 모든 산술 연산이 가능
  * 같은 위치(인덱스)의 원소들끼리(element-wise) 산술 연산을 수행.

```python
number_list1 = [1, 2, 3, -1, -2, -3]
print(type(number_list1))
print(number_list1)       # > list는 아이템과 아이템을 comma(,)로 구분해서 출력
```

```python
number_array1 = np.array([1, 2, 3, -1, -2, -3])
print(type(number_array1))
print(number_array1)     # > ndarray는 아이템과 아이템을 공백으로 구분해서 출력
```

```python
number_list2 = [10, 20, 30]
print(number_list1 + number_list2)
print(number_list2 * 3)
```

```python
number_array2 =np.ones(shape=(6), dtype=np.int64) # = np.array([1, 1, 1, 1, 1, 1])
print(number_array2)
print(number_array1 + number_array2)
```

```python
print(number_array1 * 3)
```

2차원 list vs 2차원 배열 인덱스 사용 방법

```python
# Python의 list 객체
list_2d = [
    [1, 2, 3],
    [4, 5, 6]
]

print(list_2d)
print(list_2d[0])
print(list_2d[0][0])   # 2차원 list: list[i][j]
```

```python
# numpy.ndarry 객체
array_2d = np.array(list_2d)

print(array_2d)
print(array_2d[0])
print(array_2d[0][0])  # 2차원 배열(array): array[i][j]
print(array_2d[0, 0])  # 2차원 배열(array): array[i, j]
```

## 반복문을 사용한 subplot

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 8))
print(axes)

titles = np.array([['sex', 'smoker'], ['day', 'time']])
# print(titles)
for i, row in enumerate(titles):
  for j, x in enumerate(row):
    frequency = tips[x].value_counts()
    axes[i][j].barh(y=frequency.index, width = frequency)
    axes[i, j].set_title(x)

plt.show()
```

# Scatter plot(산점도 그래프)

* 두 변수(컬럼) 간의 상관 관계를 알고 싶을 때 사용
* `y = x` 관계 시각화
  * y:  관심 변수, 종속 변수
  * x:  독립 변수

```python
# tip ~ total_bill 관계
plt.scatter(x=tips.total_bill, y=tips.tip,     # x축, y축에 사용할 컬럼(변수)
            alpha=0.5,                         # alpha: 투명도(0: 투명 / 1: 불투명)
            s=50,                              # s: size -> 마커(점)의 크기
            c='indianred',                     # c: color -> 마커(점)의 색
            marker = "o")                      # marker: 마커의 모양 (기본값: o -> 동그라미) / s, v, ^, *, +, ...
plt.xlabel('total bill')
plt.ylabel('tip')
plt.grid()
plt.show()
```

* tip ~ total_bill 상관 관계
* 남성인 경우에는 마커의 색을 'dodgerblue'로, 여성인 경우에는 'indian red'

```python
male_df = tips[tips.sex == 'Male']
female_df = tips[tips.sex == 'Female']

plt.scatter(x=male_df.total_bill, y=male_df.tip, c='dodgerblue', alpha=0.5, label='Male')
plt.scatter(x=female_df.total_bill, y=female_df.tip, c='indianred', alpha=0.5, label='Female')
plt.xlabel('total bill')
plt.ylabel('tip')
plt.grid()
plt.legend()   # legend(): scatter함수의 label parameter 값을 범례로 표시

plt.show()
```

# Line plot(선 그래프)

주로 시계열 데이터(주식 가격, 환율, 미세먼지 농도, ...)를 시각화

```python
# 선 그래프를 그리기 위한 가상 데이터
np.random.seed(1)  ## 실행할 때마다 항상 같은 난수들이 만들어지도록 하기 위해서
df = pd.DataFrame(data = {
    'day': np.arange(1, 11),     ## np.arange(start, stop, step): [start, stop) 범위의 step만큼씩 증감하는 정수들의 배열(ndarray)
    'price': np.random.rand(10)  ## np.random.rand(n): [0, 1) 범위의 실수 난수 n개를 갖는 배열(ndarray)
})
df
```

```python
# x축에 day, y축에 price로 하는 선 그래프
plt.plot(df.day, df.price, 'bo--')  ## plt.plot(x, y, fmt)
### plot의 세번째 argument fmt(format):
### -> color(색상), marker(마커 모양), linestyle(선 스타일) 정보를 문자열로 작성
### ->> color: b(blue), r(red), g(green), k(black), ...
### ->> marker: o, s, ^, v, *, +, ...
### ->> linestyle: -(직선), --(dashed line), :(doted line-점선), ...

plt.grid()
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```

```python
male_smoker = tips[(tips.sex == 'Male') & (tips.smoker == 'Yes')]       # 남성 흡연자
male_nosmoker = tips[(tips.sex == 'Male') & (tips.smoker == 'No')]      # 남성 비흡연자
female_smoker = tips[(tips.sex == 'Female') & (tips.smoker == 'Yes')]   # 여성 흡연자
female_nosmoker = tips[(tips.sex == 'Female') & (tips.smoker == 'No')]  # 여성 비흡연자

plt.scatter(x=male_smoker.total_bill, y=male_smoker.tip, c='dodgerblue', alpha=0.5, label='Male smoker', marker='*')
plt.scatter(x=male_nosmoker.total_bill, y=male_nosmoker.tip, c='dodgerblue', alpha=0.5, label='Male nosmoker')
plt.scatter(x=female_smoker.total_bill, y=female_smoker.tip, c='indianred', alpha=0.5, label='Female smoker', marker='*')
plt.scatter(x=female_nosmoker.total_bill, y=female_nosmoker.tip, c='indianred', alpha=0.5, label='Female nosmoker')

plt.grid()
plt.legend()
plt.xlabel('total bill')
plt.ylabel('tip')

plt.show()
```

```python
# 반복문을 이용한 방법
gender = tips.sex.unique()
smoke = tips.smoker.unique()
colors = ['red', 'blue']
markers = ['o', '^']

for g, c in zip(gender, colors):
  for s, m in zip(smoke, markers):
    df = tips[(tips.sex == g) & (tips.smoker == s)]
    plt.scatter(x = df.total_bill, y = df.tip, c=c, marker=m, alpha=0.4, label=f'{g} Smoker-{s}')

plt.legend()
plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
setosa = iris[iris.species == 'setosa']
versicolor = iris[iris.species == 'versicolor']
virginica = iris[iris.species == 'virginica']

plt.scatter(x=setosa.petal_length, y=setosa.petal_width, c='black', label='setosa', alpha=0.5)
plt.scatter(x=versicolor.petal_length, y=versicolor.petal_width, c='blue', label='versicolor', alpha=0.5)
plt.scatter(x=virginica.petal_length, y=virginica.petal_width, c='red', label='virginica', alpha=0.5)
plt.grid()
plt.legend()

plt.show()
```

```python
# 반복문을 이용한 방법
spe = iris.species.unique()
col = ['black', 'blue', 'red']

for s, c in zip(spe, col):
  df = iris[iris.species == s]
  plt.scatter(x = df.petal_length, y = df.petal_width, c = c, label = f'{s}', alpha=0.5)

plt.grid()
plt.legend()
plt.xlabel('petal_length')
plt.ylabel('petal_width')

plt.show()
```
