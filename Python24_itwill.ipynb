# 필요 패키지 import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 데이터프레임 생성

```python
file_path = 'https://github.com/JakeOh/202504_itwill_oracle88/raw/refs/heads/main/apt_202504.csv'
```

```python
apt = pd.read_csv(file_path,
                  skiprows = 15,         # 첫 15줄은 파일에 대한 설명 -> skip
                  encoding='euc-kr',
                  thousands=',')         # '가격' 컬럼은 숫자 타입이어야 함!
## skiprows 파라미터: 파일에서 생략하고 넘어가는 줄 수
## encoding 파라미터: 텍스트 파일의 인코딩 타입을 설정. ex) : utf-8(Linux, MacOS), cp949(한글 Windows), euc-kr, ...
## thousands 파라미터: 숫자 자료에서 천 자리 구분자가 사용되고 있을 때 ex) 10,000
```

```python
apt['year'] = apt['계약년월'] // 100
apt['month'] = apt['계약년월'] % 100
```

```python
apt.head()
```

## 서울시 구별 아파트 매매가 분석

```python
# seoul_apt = apt[apt.province == '서울특별시'] <- 이거도 가능
# seoul_apt = apt[apt['province']] <- 이거도 가능
seoul_apt = apt[apt.province.str.startswith('서울')]
seoul_apt.head()
```

```python
seoul_apt.tail()
```

```python
# 구별 매매건수, 평균, 최댓값, 최솟값
## seoul_apt.groupby(by='county')['거래금액(만원)'].describe()  -> 이건 기술통계량 전부
seoul_apt.groupby(by='county')['거래금액(만원)'].agg(['count', 'mean', 'max', 'min'])
```

```python
result = seoul_apt.pivot_table(values='거래금액(만원)', index='county', aggfunc=['count', 'mean', 'max', 'min'])
result
```

서울에서 거래건수 상위 10개 구의 거래금액 평균, 최솟/최댓값

```python
top10_count = result.sort_values(by=(('count', '거래금액(만원)')), ascending=False).head(10)
top10_count
```

```python
sns.barplot(data=top10_count, x=('count', '거래금액(만원)'), y=top10_count.index, hue=('count', '거래금액(만원)'))
plt.show()
```

서울에서 구별 아파트 실거래가 평균 상위 10개 구의 거래건수, 거래 금액의 평균/최솟값/최댓값

```python
top10_mean = result.sort_values(by=(('mean', '거래금액(만원)')), ascending=False).head(10)
top10_mean
```

```python
sns.barplot(data=top10_mean, x=('mean', '거래금액(만원)'), y=top10_mean.index, hue=('mean', '거래금액(만원)'))
plt.show()
```

서울에서 구별 아파트 실거래가 평균 하위 10개 구의 거래건수, 거래 금액의 평균/최솟값/최댓값
