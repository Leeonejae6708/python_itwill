## indexing, slicing

`iloc` vs `loc`
* `iloc`: 정수 인덱스 기반으로 원소를 참조. list의 인덱스를 사용하는 방법과 동일  ->> 사물함 몇 번째!
* `loc`: 축 레이블 기반으로 원소를 참조  ->> 사물함 이름표!

```python
s.iloc[0]    # Series의 첫 번째 원소 참조(선택)
```

```python
s.iloc[-1]   # Series의 마지막 원소 선택
```

```python
s.iloc[:2]   # slicing의 시작 인덱스 0은 생략 가능. 첫 번째 원소부터 두 번째 원소까지 slicing
```

```python
s.iloc[-2:]  # 마지막 2개 원소를 slicing. 음수 인덱스 사용 가능. 마지막 인덱스는 생략 가능
```

```python
s.loc[0]  # Series에서 레이블이 0인 원소를 참조
```

```python
s.loc[3]  # Series에서 레이블이 -3인 원소를 참조
```

```python
# s.loc[-1] -> ValueError, KeyError 발생 ->> 리스트처럼 자동으로 부여되는 음수 인덱스는 없음
```

```python
s.loc[:2]  # loc 속성을 사용할 때도 시작 인덱스(레이블)은 생략 가능. 마지막 인덱스(레이블)의 원소도 포함!
```

## boolean indexin

* `df[조건식]` = `df.loc[조건식]`   -> 같은 코드
* 조건식을 만족하는 모든 행들을 선택

```python
# 서울 데이터만 선택
df[df['city'] == '서울']
```

```python
# 가장 최신 자료 -> 2024년 데이터만 선택
df[df['year'] == 2024]
```

Series, Dataframe에서 boolean indexing을 사용할 때 주의할 점!
* `and`, `or`, `not` 키워드 사용 불가
* 대신, `&`, `|`, `~` 연산자로 사용
* 반드시 ()를 사용해서 논리 연산의 순서를 명시해야 함

```python
# 2022, 2023년 데이터 선택
# df[(df['year'] == 2023) | (df['year'] == 2022)]
df[(df.year == 2023) | (df.year == 2022)]
```

```python
# 도시가 서울이거나 pop의 값이 1.2 이상인 데이터 선택
df[(df['city'] == '서울') | (df['pop'] >= 1.2)]
# df[(df.city == '서울') | (df.pop >= 1.2)]  -> 이건 사용 불가.
```
