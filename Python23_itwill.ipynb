# DataFrame Merge(Join)

```python
emp = pd.DataFrame(data={
    'empno':[100, 101, 200, 201],
    'ename':['홍길동', '오쌤', 'Scott', 'King'],
    'deptno':[10, 10, 20, 40]
})
emp
```

```python
dept = pd.DataFrame(data={
    'deptno':[10, 20, 30],
    'dename':['IT', 'HR', 'Sales']
})
dept
```

* `pd.merge(df1, df2, how, on, ...)` 함수(Function)
* `pd.DataFrame.merge(df, how, on, ...)` 메서드(Method)
*`how` 파라미터: merge(join)의 방식을 설정
  * inner, left, right, outer
  * merge의 기본값은 inner -> `how='inner`는 생략 가능

* `on`: merge(join)의 기준이 되는 컬럼(컬럼들의 리스트)을 설정
  * merge(join)를 하려는 2개의 DataFrame에서 컬럼 이름이 같은 경우에는 on 파라미터 설정을 생략 가능  
  * 컬럼 이름이 같은 것이 2개 이상일 경우 생략 불가

## inner join

```python
pd.merge(emp, dept)
# how='inner', on='deptno' 생략
```

```python
emp.merge(dept)
# how='inner', on='deptno' 생략
```

```python
dept.merge(emp)
```

## left (outer) join

```python
pd.merge(emp, dept, how='left')
```

```python
emp.merge(dept, how='left')
```

```python
dept.merge(emp, how='left')
```

rigth (outer) join

```python
pd.merge(emp, dept, how='right')
```

```python
emp.merge(dept, how='right')
```

```python
dept.merge(emp, how='right')
```

## (full) outer join

```python
pd.merge(emp, dept, how='outer')
```

```python
emp.merge(dept, how='outer')
```

# DataFrame 모양 바꾸기(Reshape)

## `melt`

`pd.DataFrame.melt()` 메서드: wide-form 데이터를 long-form 데이터로 변환

* `id_vars` 파라미터: melting 동작을 할 때 컬럼으로 유지할 컬럼 이름
  * `id_vars`로 설정하지 않은 모든 변수(컬럼)들은 `variable`컬럼으로 melting됨
  * `id_vars`로 설정하지 않은 모든 변수(컬럼)들의 셀에 있는 값들은 `value` 컬럼으로 melting됨
* `var_name` 파라미터: `variable` 컬럼의 이름으로 사용할 문자열
* `value_name` 파라미터: `value` 컬럼의 이름으로 사용할 문자열

```python
df = pd.DataFrame({
    'gender':['Female', 'Male'],
    'lunch': [10, 5],
    'dinner': [20, 30]
})
df
```

```python
df.melt(id_vars=['gender'])
```

```python
df.melt(id_vars=['gender'], var_name='time', value_name='size')
```

## `pivot`

`pd.DataFrame.pivot()` 메서드: long-form(세로로 긴) 데이터를 wide-form(가로로 긴) 데이터프레임으로 변환

* `index` 파라미터: pivoting한 데이터프레임에서 row index로 사용할 컬럼(들)의 이름
* `columms` 파라미터: pivoting한 데이터프레임에서 컬럼 이름으로 사용할 컬럼(들)의 이름
* `values` 파라미터: pivoting한 데이터프레임의 각 셀에 채울 값들이 있는 컬럼(들)의 이름

```python
df2 = pd.DataFrame({
    'time':['Lunch'] * 3 + ['Dinner'] * 3,
    'day': ['Fri', 'Sat', 'Sun'] * 2,
    'total_bill': np.arange(10, 70, 10),
    'tip': np.arange(1, 7)
})
df2
```

```python
df2.pivot(index='time', columns='day', values='tip')
```

```python
result = df2.pivot(index='time', columns='day', values=['total_bill', 'tip'])
result
```

```python
result.columns
```

```python
df2.pivot(index='day', columns='time', values='tip')
```

```python
df2.pivot(index='day', columns='time', values=['total_bill', 'tip'])
```

# `groupby` vs `pivot_table`

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## group by

```python
# 성별 팁의 평균
tips.groupby(by='sex', observed=True).tip.mean()
```

```python
# 성별 흡연여부별 팁의 평균
tips.groupby(by=['sex','smoker'], observed=True).tip.mean()
```

```python
# 성별 흡연여부별 tip/total_bill 평균
tips.groupby(by=['sex','smoker'], observed=True)[['tip', 'total_bill']].mean()
```

```python
# 성별 흡연여부별 tip/total_bill 평균
result=tips.groupby(by=['sex','smoker'], observed=True).mean('tip', 'total_bill')
result
```

```python
result = result.reset_index()
result
```

```python
result.pivot(index='sex', columns='smoker', values=['tip', 'total_bill'])
```

## pivot table

`pd.DataFrame.pivot_table()` 메서드 파라미터:

* `values`: pivoting을 해서 각 셀에 값으로 설정할 변수(들)
* `index`: pivoting을 할 때 row index로 설정할 변수(들)
* `columms`: pivoting을 할 때 컬럼 이름으로 설정할 변수(들)
* `aggfunc`: aggregation finction(짐계 함수). 통계 값을 계산하는 함수 이름 또는 함수 객체
  * aggfunc 파라미터의 기본값은 `mean`
  

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex', observed=True)   # aggfunc='mean' 생략 / columms도 생략 -> 컬럼은 어차피 1개이니
```

```python
# 성별 흡연여부별 팁의 평균
tips.pivot_table(values='tip', index=['sex','smoker'], observed=True)
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker', observed=True)
```

```python
# 성별 흡연여부별 팁/영수증 평균
tips.pivot_table(values=['tip', 'total_bill'], index=['sex','smoker'], observed=True)
```

```python
tips.pivot_table(values=['total_bill', 'tip'], index='sex', columns='smoker', observed=True)
```

```python
tips.pivot_table(values=['total_bill', 'tip'], index='smoker', columns='sex', observed=True)
```

성별, 흡연여부별, 요일별, 시간별 팁/영수증 평균을 계산하시오.
(성별/흡연여부는 행 인덱스로 사용하고 요일/시간은 컬럼 인덱스로 사용하시오)

```python
tips.pivot_table(values=['tip', 'total_bill'], index=['sex', 'smoker'], columns=['day', 'time'], observed=True)
```

집계 함수(aggfunc)가 여러 개인 함수

```python
# 성별 요일별 팁의 평균, 최솟값, 최댓값
tips.pivot_table(values='tip', index=['sex', 'day'], aggfunc=['mean', 'max', 'min'], observed=True)
```

```python
tips.pivot_table(values='tip', index='sex', columns='day', aggfunc=['mean', 'max', 'min'], observed=True)
```
